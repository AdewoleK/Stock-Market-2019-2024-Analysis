TREND ANALYSIS

```python
# Calculate Simple Moving Average (SMA) - 50-day and 200-day
df['SMA_50'] = df['Bitcoin_Price'].rolling(window=50).mean()
df['SMA_200'] = df['Bitcoin_Price'].rolling(window=200).mean()
```

```python
# Calculate exponential moving average (EMA)
df['EMA_50'] = df['Bitcoin_Price'].ewm(span=50, adjust=False).mean()
df['EMA_200'] = df['Bitcoin_Price'].ewm(span=200, adjust=False).mean()
```

```python
# Calculate exponential moving average (EMA)
df['EMA_50'] = df['Bitcoin_Price'].ewm(span=50, adjust=False).mean()
df['EMA_200'] = df['Bitcoin_Price'].ewm(span=200, adjust=False).mean()
```

```python
# Calculate Bollinger Bands
df['Upper_BB'] = df['Bitcoin_Price'].rolling(window=20).mean() + 2*df['Bitcoin_Price'].rolling(window=20).std()
df['Lower_BB'] = df['Bitcoin_Price'].rolling(window=20).mean() - 2*df['Bitcoin_Price'].rolling(window=20).std()
```

```python
# Calculate MACD
df['MACD'] = df['Bitcoin_Price'].ewm(span=12, adjust=False).mean() - df['Bitcoin_Price'].ewm(span=26, adjust=False).mean()
df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
```

```python
df.set_index('Date', inplace=True)
```

```python
# Plot the data
plt.figure(figsize=(12,6))
plt.plot(df['Bitcoin_Price'], label='Bitcoin Price')
plt.plot(df['SMA_50'], label='SMA 50')
plt.plot(df['SMA_200'], label='SMA 200')
plt.plot(df['EMA_50'], label='EMA 50')
plt.plot(df['EMA_200'], label='EMA 200')
plt.plot(df['Upper_BB'], label='Upper BB')
plt.plot(df['Lower_BB'], label='Lower BB')
plt.plot(df['MACD'], label='MACD')
plt.plot(df['Signal'], label='Signal')
plt.legend(loc='upper left')
plt.show()
```

SMA: If the 50-day moving average crosses above the 200-day moving average (Golden Cross), it's a bullish signal. If the 50-day moving average crosses below the 200-day moving average (Death Cross), it's a bearish signal.

When the price touches the upper band, the stock is considered overbought, and when it touches the lower band, its considered oversold.

When the MACD crosses above the Signal Line, its considered a buy signal. When it crosses below, its a sell signal.

PORTFOLIO OPTIMIZATION

```python
commodities = ['Natural_Gas_Price', 'Crude_oil_Price', 'Copper_Price','Platinum_Price', 'Silver_Price', 'Gold_Price']  # Example stocks
```

```python
# Calculate daily returns
returns = df[['Natural_Gas_Price', 'Crude_oil_Price', 'Copper_Price','Platinum_Price', 'Silver_Price', 'Gold_Price']].pct_change()

# Note most of the commodities stocks trade 252 days in a year
```

```python
# Calculate mean returns and covariance matrix
mean_returns = returns.mean()
cov_matrix = returns.cov()
```

```python
# Portfolio performance: return and volatility
def portfolio_performance(weights, mean_returns, cov_matrix):
    portfolio_return = np.sum(mean_returns * weights) * 252
    portfolio_stddev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix * 252, weights)))
    return portfolio_return, portfolio_stddev
```

```python
# Function to minimize (negative Sharpe Ratio)
def neg_sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate=0.01):
    p_return, p_stddev = portfolio_performance(weights, mean_returns, cov_matrix)
    return -(p_return - risk_free_rate) / p_stddev
```

```python
# Constraints: sum of weights = 1
constraints = {'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1}
bounds = tuple((0, 1) for _ in range(len(commodities)))  # Weights between 0 and 1
```

```python
# Run the optimization

from scipy.optimize import minimize

def optimize_portfolio(mean_returns, cov_matrix):
    num_assets = len(mean_returns)
    initial_guess = num_assets * [1. / num_assets]

    optimized_result = minimize(neg_sharpe_ratio, initial_guess, args=(mean_returns, cov_matrix),
                                method='SLSQP', bounds=bounds, constraints=constraints)

    return optimized_result.x
```

```python
# Get the optimal weights
optimal_weights = optimize_portfolio(mean_returns, cov_matrix)
print(f"Optimal Weights: {optimal_weights}")
```

```python
# Calculate Efficient Frontier

def calculate_efficient_frontier(mean_returns, cov_matrix, num_portfolios=10000):
    results = np.zeros((3, num_portfolios))
    weights_record = []

    for i in range(num_portfolios):
        weights = np.random.dirichlet(np.ones(len(commodities)), size=1)[0]
        portfolio_return, portfolio_stddev = portfolio_performance(weights, mean_returns, cov_matrix)
        results[0,i] = portfolio_stddev
        results[1,i] = portfolio_return
        results[2,i] = (portfolio_return - 0.01) / portfolio_stddev  # Sharpe Ratio
        weights_record.append(weights)

    return results, weights_record
```

```python
# Generate portfolios for Efficient Frontier
results, weights_record = calculate_efficient_frontier(mean_returns, cov_matrix)
```

```python
# Step 6: Plot the Efficient Frontier
plt.figure(figsize=(10, 7))
plt.scatter(results[0,:], results[1,:], c=results[2,:], cmap='YlGnBu', marker='o')
plt.title('Efficient Frontier')
plt.xlabel('Volatility (Standard Deviation)')
plt.ylabel('Return')
plt.colorbar(label='Sharpe Ratio')
```

```python
# Plot the optimal portfolio
opt_return, opt_stddev = portfolio_performance(optimal_weights, mean_returns, cov_matrix)
plt.scatter(opt_stddev, opt_return, marker='*', color='r', s=200, label='Optimal Portfolio')
plt.legend(labelspacing=0.8)
plt.show()
```

```python
# Print the results
optimal_return, optimal_stddev = portfolio_performance(optimal_weights, mean_returns, cov_matrix)
print(f'Optimal Weights: {optimal_weights}')
print(f'Expected Return: {optimal_return:.2f}')
print(f'Expected Volatility: {optimal_stddev:.2f}')
```
